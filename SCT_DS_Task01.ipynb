```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/SCT_TASK/world_population task_1.csv')
df.head()
```

```python
# @title Continent vs 2015 Population

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['Continent'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='2015 Population', y='Continent', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title Continent vs 2022 Population

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['Continent'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='2022 Population', y='Continent', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title Continent vs Rank

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(df['Continent'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='Rank', y='Continent', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'Country/Territory' is the column you want to analyze
column_to_analyze = 'Country/Territory'  # Change to the relevant column

# Creating a Bar Chart for Country/Territory Distribution
plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
counts = df[column_to_analyze].value_counts()
plt.barh(counts.index, counts.values, color=plt.cm.Dark2(range(len(counts))))
plt.title(f'{column_to_analyze} Distribution')
plt.xlabel('Count')
plt.ylabel(column_to_analyze)
plt.gca().spines[['top', 'right']].set_visible(False) # Remove top and right spines
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```
